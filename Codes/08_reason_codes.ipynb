```python
df_val = pd.read_csv('validation.csv',usecols=numerical+categorical+['Id'],header=0)
```

```python
df_val.set_index('Id',inplace=True)
```

```python
df_orig = df_val.copy()
```

```python
values={}
for col in categorical:
    values[col]='Unknown'
df_val.fillna(values,inplace=True)
```

```python
df_val[categorical].head()
```

```python
for col in categorical:
    if len(conversion_dict[col])==1:
        category = conversion_dict[col][0]
        df_val[col+'_dum_'+str(category)] = 0
        df_val.loc[df_val[col]==category,col+'_dum_'+str(category)]=1
    else:
        total_categories = len(conversion_dict[col])
        dummies = len(str(int(bin(total_categories)[2:],10)))
        bin_conv=[]
        for i in range(total_categories):
            bin_conv.append(conversion_dict[col][i][1])
        for j in range(dummies):
            df_val[col+'_dum_'+str(j)]=0
            for i,cat in enumerate([conv[0] for conv in conversion_dict[col]]):
                df_val.loc[df_val[col]==cat,col+'_dum_'+str(j)]=bin_conv[i]%10
                bin_conv[i]=bin_conv[i]//10
    df_val.drop(col,axis=1,inplace=True)
    print(col+' done')   
```

```python
df_original =df_val.copy()
```

```python
transform_dict = {'log':lambda x: np.log(x),'sqr':lambda x: x**2,'sqrt':lambda x: np.sqrt(x),'exp':lambda x:np.exp(x),
                 'cube':lambda x: x**3,'cuberoot': lambda x: cbrt(x)}
```

```python
os.chdir('../Statistics')
```

```python
edd = pd.read_csv('edd_v05.csv',header=0)
```

```python
def transform(x,function):
    if x is not None:
        return function(x)
    else:
        return np.nan
```

```python
for col in numerical:
    if edd[(edd['Var']==col)&(edd['Status'].notnull())].shape[0]!=0:
        function = transform_dict[edd.loc[edd['Var']==col,'Status'].values[0]]
        df_val[col] = df_val[col].apply(lambda x: transform(x,function))
```

```python
os.chdir('../Imputation_models')
```

```python
for col in [x for x in numerical if x!='SalePrice']:
    model = pickle.load(open(col+'_impute.pickle','rb'))
    if df_val[col].isnull().any():
        indices = df_val.loc[df_val[col].isnull()].index.tolist()
        df_val.loc[indices,col]=np.array(model.predict(np.array(df_val.loc[indices,imputation_cols])))
        del indices
    del model
    print(col+' imputed')
```

```python
def reason_code_lr_inferential(model,data,target='SalePrice'):
    predictors = data.drop([target],axis=1).columns
    df_reasons = pd.DataFrame()
    bias = np.array([model.intercept_ for i in range(data.shape[0])])
    df_reasons['bias'] = bias
    df_reasons.index = data.index
    df_reasons['Actual_Value'] = data['SalePrice']
    df_reasons['Predicted_Value']=model.predict(scale.transform(np.array(data[model_cols])))
    x = scale.transform(np.array(data[model_cols]))
    df_reasons['Difference'] = df_reasons['Predicted_Value']-df_reasons['Actual_Value']
    for i,col in enumerate(model_cols):
        df_reasons[col] = model.coef_[i]*x[:,i]
    return df_reasons
```

```python
def reason_code_lr_predictive(model,data,pca_weights,target='SalePrice'):
    predictors = data.drop([target],axis=1).columns
    df_reasons = pd.DataFrame()
    bias = np.array([model.intercept_ for i in range(data.shape[0])])
    df_reasons['bias'] = bias
    df_reasons.index = data.index
    df_reasons['Actual_Value'] = data['SalePrice']
    df_reasons['Predicted_Value']=model.predict(pca.transform(scale.transform(np.array(data[model_cols]))))
    x = scale.transform(np.array(data[model_cols]))
    df_reasons['Difference'] = df_reasons['Predicted_Value']-df_reasons['Actual_Value']
    for i,col in enumerate(model_cols):
        df_reasons[col] = sum([model.coef_[j-1]*pca_weights.loc['pc'+str(j),col] for j in range(1,pca.n_components+1)])*x[:,i]
    return df_reasons
```

```python
def reason_code_rf_predictive(model,data,pca_magnitude_weights,target='SalePrice'):
    predictors = data.drop([target],axis=1).columns
    df_reasons = pd.DataFrame()
    prediction, bias, contributions = ti.predict(model,pca.transform(scale.transform(np.array(data[model_cols]))))
    df_reasons['bias'] = bias
    df_reasons.index = data.index
    df_reasons['Actual_Value'] = data['SalePrice']
    df_reasons['Predicted_Value']=prediction
    x = pca.transform(scale.transform(np.array(data[model_cols])))
    df_reasons['Difference'] = df_reasons['Predicted_Value']-df_reasons['Actual_Value']
    for i,col in enumerate(model_cols):
        df_reasons[col] = np.array(np.sum(np.array([contributions[:,j-1]*pca_magnitude_weights.loc['pc'+str(j),col]
                                                   for j in range(1,pca.n_components+1)]),axis=0))
    return df_reasons
```

```python
def reason_code_gbm_inferential(model,data,y_mean=180028.44252254,target='SalePrice'):
    predictors = data.drop([target],axis=1).columns
    df_reasons = pd.DataFrame()
    learning_rate=model.learning_rate
    trees = model.estimators_[:,0]
    bias=[]
    contributions=[]
    x= scale.transform(np.array(data[model_cols]))
    for j in range(x.shape[0]):
        bias_instance=0
        prediction_final=0
        contribution_instance = np.array([0 for i in range(data[model_cols].shape[1])])
        for i,tree in enumerate(trees):
            prediction,bias_tree,contribution_tree = ti.predict(tree,x[j,:].reshape(1,-1))
            bias_instance+=bias_tree[0]*learning_rate
            contribution_instance=contribution_tree*learning_rate+contribution_instance
            prediction_final+=learning_rate*prediction
        bias.append(bias_instance)
        contributions.append(list(contribution_instance))
    contributions = np.array(contributions)
    contributions = contributions[:,0,:]
    print(contributions.shape)
    df_reasons['bias'] = np.array(bias)+y_mean
    df_reasons.index = data.index
    df_reasons['Actual_Value'] = data['SalePrice']
    df_reasons['Predicted_Value']=model.predict(scale.transform(np.array(data[model_cols])))
    x = scale.transform(np.array(data[model_cols]))
    df_reasons['Difference'] = df_reasons['Predicted_Value']-df_reasons['Actual_Value']
    for i,col in enumerate(model_cols):
        df_reasons[col] = contributions[:,i]
    return df_reasons
```
